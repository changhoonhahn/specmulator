```python
from pyDOE import lhs
import pynolh as pyNOLH
import lhsmdu as lhsMDU
```

```python
import numpy as np
import random

import matplotlib as mpl 
import matplotlib.pyplot as plt
mpl.rcParams['text.usetex'] = True
mpl.rcParams['font.family'] = 'serif'
mpl.rcParams['axes.linewidth'] = 1.5
mpl.rcParams['axes.xmargin'] = 1
mpl.rcParams['xtick.labelsize'] = 'x-large'
mpl.rcParams['xtick.major.size'] = 5
mpl.rcParams['xtick.major.width'] = 1.5
mpl.rcParams['ytick.labelsize'] = 'x-large'
mpl.rcParams['ytick.major.size'] = 5
mpl.rcParams['ytick.major.width'] = 1.5
mpl.rcParams['legend.frameon'] = False
%matplotlib inline
```

Testing pyDOE LHD 

```python
A = lhs(5, samples=17, criterion='cm', iterations=1000)
```

```python
print A.shape
```

```python
fig = plt.figure(figsize=(9,9))
for i in range(A.shape[1]):
    for j in range(A.shape[1]): 
        if i < j:
            sub = fig.add_subplot(A.shape[1],A.shape[1],A.shape[1]*j+i+1)
            sub.scatter(A[:,i], A[:,j])
            sub.set_xlim([0., 1.])
            sub.set_xticks([0., 0.5, 1.])
            sub.set_xlabel(r'$\theta_'+str(i)+'$')
            sub.set_ylim([0., 1.])
            sub.set_yticks([0., 0.5, 1.])            
            sub.set_ylabel(r'$\theta_'+str(j)+'$')
```
